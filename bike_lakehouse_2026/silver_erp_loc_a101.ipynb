```python
import pyspark.sql.functions as F
from pyspark.sql.types import StringType, DateType
from pyspark.sql.functions import trim, col
```

## Read Bronze table

```python
df = spark.table("workspace.bronze.erp_loc_a101")
```

```python
df.display()
```

```python
col_counts = df.select([F.count(col(c)).alias(c) for c in df.columns])
display(col_counts)
```

```python
null_counts = df.select([F.count(F.when(col(c).isNull(), c)).alias(c) for c in df.columns])
display(null_counts)
```

## handle Null

```python
from pyspark.sql import Window

mode_gen = df.groupBy("CNTRY").count().orderBy(F.desc("count")).first()["CNTRY"]
df = df.withColumn(
    "CNTRY",
    F.when(col("CNTRY").isNull(), mode_gen).otherwise(col("CNTRY"))
)
display(df)
```

## Trimming

```python
for field in df.schema.fields:
    if isinstance(field.dataType, StringType):
        df = df.withColumn(field.name, trim(col(field.name)))
```

## Customer ID Cleanup

```python
df = df.withColumn("cid", F.regexp_replace(col("cid"), "-", ""))
```

```python
df.display()
```

## Country Normalization

```python
df = df.withColumn(
    "cntry",
    F.when(col("cntry") == "DE", "Germany")
     .when(col("cntry").isin("US", "USA"), "United States")
     .when((col("cntry") == "") | col("cntry").isNull(), "n/a")
     .otherwise(col("cntry"))
)
```

## Renaming Columns

```python
RENAME_MAP = {
    "cid": "customer_number",
    "cntry": "country"
}
for old_name, new_name in RENAME_MAP.items():
    df = df.withColumnRenamed(old_name, new_name)
```

```python
df.display()
```

## Sanity checks of dataframe

```python
df.limit(10).display()
```

## Writing Silver Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("workspace.silver.erp_customer_location")
```
